# Data Splitting

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df.drop('ReconciliationRuleId', axis=1)
y = df['ReconciliationRuleId']

label_encoders = {}
for column in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

# Encode the target variable
le_target = LabelEncoder()
y = le_target.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=6)
```

# Model Selection and Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf_model = RandomForestClassifier(random_state=6)
```

```python
param_grid = {
    'n_estimators': [5, 10, 12, 13],
    'max_depth': [5,6,7,8,9],
    'min_samples_split': [10, 20],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': [2, 3, 4, 5, 'sqrt'],
    # 'bootstrap': [True, False],
    'class_weight': ['balanced', 'balanced_subsample']
}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(rf_model, param_grid,
                           cv=2,
                           scoring='accuracy', 
                           n_jobs=-1,  # Use all available cores
                           verbose=2)
grid_search.fit(X_train, y_train)
```

```python
# Retrieve all parameter combinations evaluated
all_params = grid_search.cv_results_['params']

# Display parameter combinations with corresponding mean test scores
for i, params in enumerate(all_params):
    print(f"Model {i}: Params = {params}, Mean Test Score = {grid_search.cv_results_['mean_test_score'][i]}, Rank Test Score = {grid_search.cv_results_['rank_test_score'][i]}")
```

```python
# best_model = RandomForestClassifier(**best_params, random_state=6)
best_model = RandomForestClassifier(class_weight= 'balanced', max_depth= 5, max_features=5, min_samples_leaf= 1, min_samples_split= 10, n_estimators= 5, random_state=6)
best_model.fit(X_train, y_train)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions
y_pred = best_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print(classification_report(y_test, y_pred))

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Feature importance

```python
feature_importances = pd.Series(best_model.feature_importances_, index=X_train.columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Important Features')
plt.show()
```

```python
# Get predicted probabilities
y_prob = best_model.predict_proba(X_test)
predictions_with_ruleids_and_probs = []

for prob in y_prob:
    sorted_indices = prob.argsort()[::-1]
    sorted_ruleids = le_target.inverse_transform(sorted_indices)
    sorted_probs = prob[sorted_indices]

    combined = [f'{ruleid} ({round(p * 100, 2)}%)' for ruleid, p in zip(sorted_ruleids, sorted_probs)]
    
    predictions_with_ruleids_and_probs.append(combined)

# Convert the list to a DataFrame
predictions_df = pd.DataFrame(predictions_with_ruleids_and_probs, columns=[f'RuleID_{i+1}' for i in range(len(le_target.classes_))])

# Show the DataFrame
predictions_df.head()
```

```python
# Convert y_test to its original labels
y_test_original = le_target.inverse_transform(y_test)

# Create a DataFrame for y_test
y_test_df = pd.DataFrame(y_test_original, columns=['True_Labels'])

# Concatenate X_test, y_test, and predictions_df
combined_df = pd.concat([X_test.reset_index(drop=True), y_test_df.reset_index(drop=True), predictions_df.reset_index(drop=True)], axis=1)

combined_df
```
